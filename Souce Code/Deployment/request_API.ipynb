```python
import matplotlib.pyplot as plt
import requests
import time
```

```python
# Load image
img = plt.imread("/content/2.jpg")

# Display image
plt.figure(dpi=130)
plt.imshow(img)
plt.show()
```

```python
url = 'https://janus-gates.up.railway.app/api/ai/predict'
image = open('/content/2.jpg', 'rb')
st=time.time()
response = requests.post(url, files={'image': image})

print(f"Prediction time without GPU is {time.time()-st} seconds")
print(response.json())
```
